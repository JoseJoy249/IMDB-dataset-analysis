```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
import time
from IPython.display import display, HTML
from collections import Counter
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
import csv
```

```python
path = '/Users/josejoy/Desktop/ECE 180 Programming for data science/project/data/'
movies = pd.read_csv(path+'Movies_21Feb.csv',index_col=0)
titles = pd.read_csv(path+'Cleanedup_titles.csv',sep = '\t',index_col=0)
people = pd.read_csv(path+'Cleanedup_people.csv',sep = '\t',index_col=0)
```

## Word cloud of movie names

```python
text = ' '.join(movies['primaryTitle'])
from nltk.corpus import stopwords
stopwords1 = list(stopwords.words('english'))
stopwords1 = [ str(x) for x in stopwords1]
# stopwords1 = set(' ')
mask1 = np.array(Image.open("im1.gif"))
wc = WordCloud(background_color="white", max_words=1000, mask=mask1, stopwords=stopwords1)
wc.generate(text)
plt.figure(figsize = (15,5))
plt.subplot(121)
plt.imshow(wc)
plt.axis("off")
plt.title('Word cloud of titles without stop words', fontsize = 15)

stopwords1 = set(' ')
mask1 = np.array(Image.open("im1.gif"))
wc = WordCloud(background_color="white", max_words=1000, mask=mask1, stopwords=stopwords1)
wc.generate(text)
plt.subplot(122)
plt.imshow(wc)
plt.axis("off")
plt.title('Word cloud of titles with stop words', fontsize = 15)

plt.show()
```

## Wordcloud of reviews

```python
path1 = '/Users/josejoy/Desktop/ECE 180 Programming for data science/project/data/'
reviews = pd.read_csv(path1+'review.csv',sep = '\t',index_col=0) # Review dataset
text = list( reviews['review contents'] )
from nltk.corpus import stopwords
stopwords = list(stopwords.words('english'))
stopwords = [ str(x) for x in stopwords]
punc = string.punctuation + '0123456789'

words = ' '.join(text).lower()
words = ''.join([w for w in words if w not in punc]).split()
text = ' '.join(words)
mask1 = np.array(Image.open("im5.png"))
wc = WordCloud(background_color="white", max_words=1000, mask=mask1, stopwords=stopwords)
wc.generate(text)
plt.figure(figsize = (12,12))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
