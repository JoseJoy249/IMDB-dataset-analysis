```python
import csv
import os
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show, output_file
```

```python
title_csv = open('dataset/Cleanedup_titles.csv','r')
title_reader = csv.reader(title_csv,delimiter='\t')

idx = 0
row = next(title_reader, None)
row = next(title_reader, None)
genre_rate = {}

while row:
    genres = row[9].split(',')
    dirs = row[13].split(',')
    for g in genres:
        if g == '\\n' or g=='NaN':
            continue
        if g not in genre_rate:
            genre_rate[g] = []
        genre_rate[g].append(float(row[10]))
    idx += 1
    row = next(title_reader, None)

title_csv.close()
```

```python
if not os.path.exists('img'):
    os.mkdir('img')
```

```python
genre_weight = [len(genre_rate[i]) for i in genre_rate]
plt.figure(figsize=(15,10))
plt.pie(genre_weight, autopct='%1.1f%%')
plt.legend(genre_rate.keys(), bbox_to_anchor=(1.2, 1))
plt.title('Pie Chart of Genres')
plt.savefig('img/genre_pie.png')
plt.show()
```

```python
plt.figure(figsize=[15,10])
plt.boxplot(genre_rate.values(), labels=genre_rate.keys(), notch=True)
plt.xticks(rotation=45)
plt.title('Box Plot of different Genre')
plt.savefig('img/genre_box.png')
plt.show()
```

```python
title_csv = open('dataset/Cleanedup_titles.csv','r')
title_reader = csv.reader(title_csv,delimiter='\t')

idx = 0
row = next(title_reader, None)
row = next(title_reader, None)
score = []
numVote = []
while row:
    if row[10]!='NaN' and row[10]!='\\n' and row[11]!='NaN' and row[11]!='\\n':
        score.append(float(row[10]))
        numVote.append(float(row[11]))
    idx += 1
    row = next(title_reader, None)

title_csv.close()
```

```python
plt.figure()
plt.scatter(numVote, score)
plt.title('numVote vs rate')
plt.savefig('img/numVote_vs_rate.png')
plt.show()
```
